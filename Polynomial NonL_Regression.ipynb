```python
#importing libraries
import numpy as np # for maths
import matplotlib.pyplot as plt   #for visuals
import pandas as pd  #for handling data sets
from sklearn.preprocessing import Imputer  #to take care of missing data
```

```python
data = pd.read_csv("C:/Users/Vishnu/Downloads/ML/Polynomial/Position_Salaries.csv")
```

```python
X= data.iloc[:,1:2].values  #so that the independent variable is taken as a matrix
Y= data.iloc[:,2].values
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
LinReg.fit(X,Y)
```

```python
#Non Linear Regression
from sklearn.preprocessing import PolynomialFeatures
#converts X to X^2, X^3 as desired

Poly_reg = PolynomialFeatures(degree=4)  #try with 2 first
X_poly = Poly_reg.fit_transform(X)

#adds a constant term automatically
```

```python
Linreg2 = LinearRegression() #to include the polynomial in our LR model
Linreg2.fit(X_poly,Y)
```

```python
#Visualizing the linear regression model
plt.scatter(X,Y,color='blue')
plt.plot(X,LinReg.predict(X),color='red')
plt.title('Linear Reg fit')
plt.xlabel('Salary Positions')
plt.ylabel('Salary')
```

```python
#Visualizing polynomial regression

#Below two lines : so that we can get rid of the straight lines in our graph and make it more smooth
X_grid = np.arange(min(X),max(X),0.1)  
X_grid = X_grid.reshape(len(X_grid),1)  #to reshape to matrix


plt.scatter(X,Y,color='blue')
plt.plot(X_grid,Linreg2.predict(Poly_reg.fit_transform(X_grid)),color='red')
plt.title('Poly_NLinear Reg fit')
plt.xlabel('Salary Positions')
plt.ylabel('Salary')
```
